```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
iris = fetch_ucirepo(id=53)

# data (as pandas dataframes)
X = iris.data.features
y = iris.data.targets

# metadata
print(iris.metadata)

# variable information
print(iris.variables)
```

```python
# 1. Librerías

import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from ucimlrepo import fetch_ucirepo
# 2. Cargar dataset Iris

iris = fetch_ucirepo(id=53)

X = iris.data.features.values
y = iris.data.targets.values.ravel()

# 3. Evaluar diferentes k con validación cruzada

resultados = {}

for k in range(1, 11):

    # Pipeline = Normalización + Modelo
    modelo = Pipeline([
        ('scaler', StandardScaler()),
        ('knn', KNeighborsClassifier(n_neighbors=k))
    ])

    # Validación cruzada (5 folds)
    scores = cross_val_score(modelo, X, y, cv=5)

    resultados[k] = scores.mean()

    print(f"k = {k} → Accuracy promedio = {scores.mean():.4f}")

# 4. Mejor k

mejor_k = max(resultados, key=resultados.get)

print("\nMejor k:", mejor_k)
print("Mejor accuracy:", resultados[mejor_k])
```

### Codo

El método del codo consiste en graficar el error o la precisión en función de k y seleccionar el punto donde la mejora comienza a disminuir significativamente, formando un “codo” en la gráfica.

Entonces…

El codo = mejor balance

Porque:

No usas un k muy pequeño ❌

No usas un k innecesariamente grande ❌

Obtienes buen rendimiento sin complicar el modelo ✅
