```python
import pandas as pd

dfc = pd.read_csv('clientes.csv')
dfo = pd.read_csv('ordenes.csv')

merged_df = pd.merge(dfc, dfo, on='cliente_id', how='left')
total_filas = len(merged_df)
monto_total_por_cliente = merged_df.groupby('cliente_id')['monto'].sum().fillna(0)
monto_cliente_1 = monto_total_por_cliente[1]

print(f"Número total de filas del left merge: {total_filas}")
print(f"El monto total por el cliente_id=1 es: {monto_cliente_1}")
```

```python
import pandas as pd

df = pd.read_csv('sensores.csv')

df = df[['ts', 'temperatura', 'humedad']]

df['ts'] = pd.to_datetime(df['ts'], format='mixed', errors='coerce')

mediana_temp = df['temperatura'].median()
media_hum = df['humedad'].mean()

df['temperatura'] = df['temperatura'].fillna(mediana_temp)
df['humedad'] = df['humedad'].fillna(media_hum)


nat_count = df['ts'].isna().sum()

mediana_temp_final = round(df['temperatura'].median(), 1)

media_hum_final = round(df['humedad'].mean(), 3)

print(f"Número de NaT en ts: {nat_count}")
print(f"Mediana final de temperatura tras imputación: {mediana_temp_final}")
print(f"Media final de humedad tras imputación: {media_hum_final}")
```
