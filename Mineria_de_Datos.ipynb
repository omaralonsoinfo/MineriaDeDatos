#### The Basics of Numpy Arrays

```python
import numpy as np
np.random.seed(0)  # seed for reproducibility

x1 = np.random.randint(10, size=6)  # One-dimensional array
x2 = np.random.randint(10, size=(3, 4))  # Two-dimensional array
x3 = np.random.randint(10, size=(3, 4, 5))  # Three-dimensional array
```

```python
print("x3 ndim: ", x3.ndim) # La propiedad ndim nos dice el numero de dimensiones
print("x3 shape:", x3.shape)
print("x3 size: ", x3.size) # 3*4*5 = 60
```

```python
print(x2.dtype) # dtype nos dice el tipo de datos
```

```python
#Itemsize nos regresa el tama침o en size
print("itemsize:", x3.itemsize, "bytes")
print("nbytes:", x3.nbytes, "bytes") #nbytes es el tama침o del elemento por el numero
```

### Estadistica B치sica

#### Media (Mean)


```python
data = np.array([1, 2, 3, 4, 5])
media = np.mean(data)
print(media)
```

#### Promedio (Average) y Promedio Ponderado


```python
data = np.array([1, 2, 3, 4, 5])
weights = np.array([1, 2, 3, 4, 5])
promedio = np.average(data)
promedio_ponderado = np.average(data, weights=weights)
```

#### Mediana


```python
mediana = np.median(data)
```

#### Varianza y Desviaci칩n Est치ndar


```python
varianza = np.var(data)
desviacion = np.std(data) #desviacion raiz cuadrada de la varianza
```

#### 游댃 Generaci칩n de Datos Aleatorios
