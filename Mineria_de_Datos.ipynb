#### Ciencia de Datos

 La ciencia de datos es un campo que combina matemáticas, estadística, programación y herramientas tecnológicas para analizar grandes cantidades de información y descubrir patrones

#### Mineria de Datos

 La minería de datos es el proceso de buscar patrones y conocimientos escondidos dentro de grandes cantidades de información, como si estuvieras buscando pepitas de oro en una mina.
