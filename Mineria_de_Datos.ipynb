```python
#comando magico timeit pára evaluar el desempeño, en terminos de tiempo
#comparar si un algoritmo es mas rapido que otro
%timeit L = [n ** 2 for n in range(1000)]
```

```python
#comando magico sirve para evaluar el desempeño, solo que este abarca varias lineas de codigo en lugar de una linea
%%timeit
L = []
for n in range(1000):
  L.append(n ** 2)
```
